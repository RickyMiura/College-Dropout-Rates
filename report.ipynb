# The Relevance of Socioeconomic Factors on College Dropout Rates

# Abstract

Our group's research discussed the relationship between students' background, mostly financial background, and student college drop-out rate and academic success. The background research shows that high-level education is often accompanied by a huge financial burden, making students resort to taking out loans. This might cause students to get an idea of giving up on pursuing their education. From tuition fees, scholarship, gender, and other specific parts, based on 4424 responses, we make data visualizations and create a machine learning model, we can conclude that students’ parents' education level can be a vector of students dropping out of college or not. Besides that scholarships or being in debt mostly impact the predictor. The financial status can influence whether a student will drop out of college.

# Research Question

Can the financial status of a student, measured through family socioeconomic status, tuition fees, student debt, and scholarship recipient, be used as a significant predictor for dropout rate and level of academic success?

## Background and Prior Work

While college mainly serves as a place for academics and obtaining a degree of higher education, it is also a space where students from vastly different backgrounds come together, all striving towards similar educational objectives. It offers the opportunity to encounter new individuals daily, each with their own unique stories and distinct backgrounds. However, the harsh reality is that while everyone aims for the same objective—a degree—not all succeed in reaching this goal. Despite the many students walking the stage to claim their diplomas, there are many who, unable to realize their hopes and dreams, find themselves forced to withdraw from their academic pursuits. When considering individuals who have dropped out of college, it's often assumed that they left due to the challenging nature of the coursework and the demanding environment of higher education. However, what if external factors also play a significant role in determining whether a student decides to drop out?

While basic educational levels like middle school and high school are generally provided at no cost to the public, the landscape shifts dramatically when it comes to college. Higher education carries a hefty price tag, making it a significant financial undertaking that can place substantial economic strain on students and their families. These financial demands often exceed what many families can comfortably afford, compelling them to resort to taking out loans. Perhaps, the daunting idea of shouldering loan repayments long after graduation could deter students, influencing some to discontinue their studies. 

To address this question, we aim to analyze the prevalence of college dropouts across a diverse range of student backgrounds. Specifically, we will investigate how the occupations and educational qualifications of students' families—which are significant indicators of a family's financial status—affect dropout rates. By examining the correlation between family financial standing and the likelihood of students discontinuing their college education, we seek to uncover the extent to which economic factors contribute to dropout decisions. In addition to examining the impact of family financial background on college dropout rates, we also plan to explore the influence of other variables, including gender and scholarship receipt, on students' decisions to leave college. 

By integrating these factors into our analysis, we aim to uncover significant correlations that highlight the impact of economic constraints and social factors on student dropout rates.

## SVM Analysis

```python
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.svm import SVC
```

```python
data = dropouts
dataY = data['Target']
dataX = data.drop(columns=['Target', 'Gender', 'Tuition fees up to date'])

n_splits = 5    

skf = StratifiedKFold(n_splits=n_splits, shuffle=True)
accuracyValid = 0
accuracyTrain = 0   

def score(predY, validY):
    #accuracy for classification problems
    correct = 0
    total = 0
    
    for i in range(len(predY)):
        if predY[i] == validY.values[i]:
            correct += 1
        total += 1
        
    return correct / total
        


for i, (train_index, test_index) in enumerate(skf.split(dataX, dataY)):

    #DATA PROCESSING
    
    print("Data Processing Fold: ", i)
    
    trainData = data.iloc[train_index]

    trainX = dataX.iloc[train_index]
    trainY = dataY.iloc[train_index]

    validX = dataX.iloc[test_index]
    validY = dataY.iloc[test_index]
        

    print("creating features")
    # One-hot encode 'Mother Occupation' and 'Father Occupation' in trainX
    X = trainX
   
    validFeatX = validX

    print("fitting model")
    
    clf = make_pipeline(StandardScaler(), SVC(gamma='auto', C = 0.1, cache_size=2000,class_weight='balanced', kernel='rbf'))
    clf.fit(X, trainY)

    print("validating")

    predValid = clf.predict(validFeatX)
    validScore= score( predValid, validY)
    print("Validation Score: ", validScore)        
    
    predTrain = clf.predict(X)
    trainScore = score( predTrain, trainY)
    
    print("Training Score: ", trainScore)
    
    accuracyValid += validScore
    accuracyTrain += trainScore
    
    print("------------------------------")
    
print("Average Validation Score: ", accuracyValid / n_splits)
print("Average Training Score: ", accuracyTrain / n_splits)
```

```python
from sklearn.inspection import permutation_importance

# Fit your model
clf = make_pipeline(StandardScaler(), SVC(gamma='auto', C = 0.1, cache_size=2000,class_weight='balanced', kernel='rbf'))
clf.fit(X, trainY)

# Calculate permutation importance
result = permutation_importance(clf, validFeatX, validY, n_repeats=10, random_state=42, n_jobs=2)

sorted_idx = result.importances_mean.argsort()[::-1]

for i in sorted_idx:
    if result.importances_mean[i] - 2 * result.importances_std[i] > 0:
        print(f"{validFeatX.columns[i]:<8}\t"
              f"{result.importances_mean[i]:.3f}"
              f" +/- {result.importances_std[i]:.3f}")
```

We decided to use an SVM model using an RBF kernel in order to capture any non-linearity in the data. The hyperparameters were tuned in order to prevent overfitting while getting the highest validation accuracy. In training we use 5-fold cross-validation and the final accuracy levels are derived from the averages of each fold. To determine which features of our data were the most impactful we implemented permutations of the model while dropping one feature at a time. Features that are deemed important are ones in which the model’s performance drops by over 2 standard deviations in the course of 50 trials.
